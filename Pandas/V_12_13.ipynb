```python
def run_scored(batsman):
    a = data[data['batsman']==batsman]
    return a.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False)
```

```python
death_over = data[data['over']>15]
death_over
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
batsman_list = all_batsman[x].index.tolist()
```

```python
final = data[data['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
sr
```
